```python
import numpy as np
import matplotlib.pyplot as plt
np.set_printoptions(threshold=np.inf)
```

```python
def sign(x):
    if x >= 0:
        x = 1
    else:
        x = -1
    return x

            
class hopfield():
    def __init__(self, size, npatterns, theta, trainX):
        self.W = np.zeros([npatterns,npatterns])
        self.size = size
        self.npatterns = npatterns
        self.theta = theta
        
        self.X = trainX
    
    def calc_weights(self):
        ro = 1/(self.size*self.npatterns)*np.sum(self.X)
#         print("ro = ",ro)
        for x in self.X:
            self.W += np.outer(x-ro,x-ro)
        for i in range(self.npatterns):
            self.W[i,i] = 0
#         print(self.W)
    
    def update_x(self, x, W):
#         print("len = ", len(x))
        for i in range(len(x[0])):
#             print("before ", np.dot(x,W[i,:]))
            x[0,i] = 0.5 + 0.5*sign(np.dot(x,W[i,:])-self.theta)
#             print("after", sign(np.dot(x,W[i,:])))
        return x

    def update_batch(self,x):
        new = np.dot(x,self.W)-self.theta
        new[new>=0] = 1
        new[new<0] = 0
        return new
    
    def update_random(self,x):
        new = x.copy()
        n_dims = x.shape[0]
        for i in range(n_dims):
            unit = np.random.randint(0,n_dims)
            xsum = 0
            for j in range(n_dims):
                xsum += np.multiply(self.W[unit][j], new[j])
            new[unit] = sign(xsum)
            
        return new
            
    def train(self):       
        self.calc_weights()            
#         self.update_x(x,self.W)
    
    def recall(self, x, n_epochs):
        for epoch in range(n_epochs):
            x = self.update_x(x, self.W)
#             print(x)
        return x

    def recall_batch(self,x, max_iter=100):
        error = 1
        iters = 0
        
        while error > 0:
            new = self.update_batch(x)
            error = sum(abs(new-x))
            x = new.copy()
            iters = iters+1
            if(iters > max_iter):
                break
                
        print("iters: " + str(iters))
        return x
    
    def recall_random(self,x, max_iter=15):
        error = 1
        iters = 0
        
        while error > 0:
#             print(iters)
            new = self.update_random(x)
            error = sum(abs(new-x))
            x = new.copy()
            iters+=1
            if(iters > max_iter):
                break
        
        print("iters: " + str(iters))
        return x
 
 
```

```python
def gen_random_data(dim, length, sparsity):
    train_X = np.random.uniform(0,1,(dim,length))
    train_X[train_X > sparsity] = 1
    train_X[train_X <= sparsity] = 0
    return train_X
```

```python
def add_noise(image, percentage_flip):
    result = np.copy(image)
    n_flip = int(len(image)*percentage_flip/100)
    units = np.random.choice(len(image),n_flip,replace=False)
    for i in units:
        result[i] = np.abs(result[i]-1)
    return result

def recall_accuracy(input_image,recall_image):
    correct = 0
    counter = 0
    for i in range(len(input_image)):
        if input_image[i] == 1:
            counter += 1
            if input_image[i] == recall_image[i]:
                    correct += 1
    return correct*100/counter
```

```python
#train with random data
theta = 0.0

dim = 55
train_X = gen_random_data(dim,1000,0.5)
c_test = train_X[0]

hopfield_net = hopfield(train_X.shape[0],train_X.shape[1],theta,train_X)
hopfield_net.train()
acc_vals = []
steps = []
for i in range(21):
    noisy_data = add_noise(c_test,i*5)
    test = hopfield_net.update_batch(noisy_data)
    acc_vals.append(recall_accuracy(c_test,test))
    steps.append(i*5)
    
plt.plot(steps,acc_vals)
plt.xlabel('% of pixels flipped')
plt.ylabel('recall accuracy')
plt.title('Change in recall accuracy with ' + str(dim) + ' stored random vectors')
plt.savefig('p1_accuracu.png')
```

```python
# def test_sparse_capacity:

theta_list = [0.0,0.3,0.6,0.9]
for theta in theta_list:
    dim = range(1,101)


    acc_vals = []
    steps = []
    for d in dim:
        avg_acc = 0
        for i in range(2):
            train_X = gen_random_data(d,1024,0.9)
            c_test = train_X[0]
            hopfield_net = hopfield(train_X.shape[0],train_X.shape[1],theta,train_X)
            hopfield_net.train()
            noisy_data = add_noise(c_test,25)
            test = hopfield_net.update_batch(noisy_data)
            avg_acc += recall_accuracy(c_test,test)/2
        acc_vals.append(recall_accuracy(c_test,test))
        steps.append(d)

    plt.plot(steps,acc_vals)
    plt.xlabel('patterns stored')
    plt.ylabel('Recall accuracy')
    plt.title('Recall accuracy for sparse (ro=0.01) patterns')
    plt.legend(["0.0","0.3","0.6",'0.9'])
    plt.savefig('qqqqqqqqqqqqq3.png')
```
